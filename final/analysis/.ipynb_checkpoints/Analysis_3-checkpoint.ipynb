# Analysis 3
Analyzed each match in premire league 2015 and calculated the conversion rate of chances created by each team home,away and in total.

```python
import requests,json,os
import re,datetime
import shutil
import matplotlib.pyplot as mp
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
pl2015=pd.read_csv('../data/pl2015.csv') #Reading the already merged and saved premire league csv.   
```

```python
home=pl2015.groupby('homeTeamName').sum()[['HST','FTHG']] #Grouping the data according to home teams.
home.reset_index(level=(0), inplace=True)
home.rename(columns={'homeTeamName':'Team'}, inplace=True)
home['HomeChances']=home['HST']+home['FTHG']     #Adding column to hold the total chances created by home team.
home['%_convHome']=(home['FTHG']/home['HomeChances'])  #Adding the column to hold conversion rate of the total chances.
home=home.sort_values(by=['HomeChances'], ascending=[True])
#home=home.sort_values(by=['%_convHome'], ascending=[False])
#home
```

```python
away=pl2015.groupby('awayTeamName').sum()[['AST','FTAG']] #Grouping the data according to away teams.
away.reset_index(level=(0), inplace=True)
away.rename(columns={'awayTeamName':'Team'}, inplace=True)
away['AwayChances']=away['AST']+away['FTAG']   #Adding column to hold the total chances created by away team.
away['%_convAway']=(away['FTAG']/away['AwayChances'])  #Adding the column to hold conversion rate of the totla chances.
away=away.sort_values(by=['AwayChances'], ascending=[True])
#away=away.sort_values(by=['%_convAway'], ascending=[False])
#away
```

```python
PL=pd.merge(home, away, on=['Team'], how='inner') #Merging both the away and home data according to team names
PL['ToTchances']=PL['HomeChances']+PL['AwayChances'] #Adding column to hold the total chances a team has created.
PL['Converted']=PL['FTHG']+PL['FTAG']
PL['%_conv']=(PL['Converted']/PL['ToTchances']) #Adding the column to hold conversion rate of the total chances.
PL=PL.sort_values(by=['ToTchances'], ascending=[True])[['Team','ToTchances','Converted','%_conv']]
#PL.sort_values(by=['%_conv'], ascending=[False])[['Team','ToTchances','Converted','%_conv']]
```

```python
#Plot for total chances created vs conversion rate for each team.
ax=sns.pointplot(x=PL['ToTchances'], y=PL['%_conv'], hue=PL['Team'])
ax.legend_.remove()
ax.grid(True)
i=0
for L, y in zip(PL['Team'], PL['%_conv']):
    ax.text(i*.850, (y-0.0015), L,color='Black', fontsize=7,
        horizontalalignment='left',rotation=-25)
    i+=1

mp.ylabel('Conversion %')
mp.xlabel('Total Chances Created')

mp.show()
```

```python
#Plot for total chances created vs conversion rate for each team at home.
ax=sns.pointplot(x=home['HomeChances'], y=home['%_convHome'], hue=home['Team'])
ax.legend_.remove()
ax.grid(True)
i=0
for L, y in zip(home['Team'], home['%_convHome']):
    ax.text(i*.850, (y-0.001), L,color='Black', fontsize=7,
        horizontalalignment='left',rotation=-25)
    i+=1

mp.ylabel('Conversion %')
mp.xlabel('Chances Created At Home')

mp.show()
```

```python
#Plot for total chances created vs conversion rate for each team at away.
ax=sns.pointplot(x=away['AwayChances'], y=away['%_convAway'], hue=away['Team'])
ax.legend_.remove()
ax.grid(True)
i=0
for L, y in zip(away['Team'], away['%_convAway']):
    ax.text(i*.788, (y-0.003), L,color='Black', fontsize=7,
        horizontalalignment='left',rotation=-25)
    i+=1

mp.ylabel('Conversion %')
mp.xlabel('Chances Created At Away')

mp.show()
```
