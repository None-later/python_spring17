```python
from email.parser import Parser
import os,csv
import matplotlib.pyplot as mp
from datetime import datetime
import string, re
dir_path='../../midterm/data/enron/maildir/'
```

```python
i=0
mail_sub={}
thread_starter={}
for root, dirs, files in os.walk(dir_path):
    if "sent" in root:
        for f in files:
            with open(os.path.join(root,f),'r',encoding='utf-8',errors='ignore') as e:
                data=e.read()
                email=Parser().parsestr(data)

                if 're:' in email["Subject"].lower() or 'fw:' in email["Subject"].lower() :
                    
                    yo=email["Subject"].lower().replace("re: ", "")
                    yo=yo.replace("re:", "")
                    yo=yo.replace("fw: ", "")
                    yo=yo.replace("fw:", "")
                    if yo and not yo.isspace():
                         
                        if yo not in mail_sub:
                            mail_sub[yo]=1
                        else:
                            mail_sub[yo]+=1
                else:
                    if email["Subject"].lower() and not email["Subject"].lower().isspace():
                         
                            
                        if email["Subject"].lower() not in mail_sub:
                            mail_sub[email["Subject"].lower()]=1
                            if email["From"] not in thread_starter:
                                thread_starter[email["From"]]=1
                            else:
                                thread_starter[email["From"]]+=1
                        elif email["Subject"].lower() in mail_sub :
                            mail_sub[email["Subject"].lower()]+=1  
```

```python
top100_mail_thread = {m: mail_sub[m] for m in sorted(mail_sub, key=mail_sub.get, reverse=True)[0:100]}
for x in sorted(top100_mail_thread, key=top100_mail_thread.get, reverse=True):
    print(x,top100_mail_thread[x])
```

```python
x1=[]
y1=[]
top50_thread_starter = {t: thread_starter[t] for t in sorted(thread_starter, key=thread_starter.get, reverse=True)[:50]}
for x in sorted(top50_thread_starter, key=top50_thread_starter.get, reverse=True):
    x1.append(x)
    y1.append(top50_thread_starter[x])
    print(x,top50_thread_starter[x])
```

```python
mp.style.use('ggplot') 
mp.style.use('seaborn-poster')
mp.bar(range(len(sorted(top50_thread_starter, key=top50_thread_starter.get, reverse=True))),y1,align='center',linewidth=2,width=.75,color='darkgreen')
mp.xticks(range(len(sorted(top50_thread_starter, key=top50_thread_starter.get, reverse=True))),x1,rotation=90)

mp.ylabel('Thread Count')
mp.xlabel('Employee Mailid')

mp.show()
```
