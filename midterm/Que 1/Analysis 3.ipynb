```python
from email.parser import Parser
import os,csv
import matplotlib.pyplot as mp
from datetime import datetime
import string, re
dir_path='../../midterm/data/enron/maildir/'
```

```python
sender={}
for root, dirs, files in os.walk(dir_path):
    if "sent" in root:
        for f in files:
            with open(os.path.join(root,f),'r',encoding='utf-8',errors='ignore') as e:
                data=e.read()
                email=Parser().parsestr(data)
                if email["From"] not in sender:
                    sender[email["From"]]=1
                else:
                    sender[email["From"]]+=1
```

```python
reciever={}
for root, dirs, files in os.walk(dir_path):    
    if "sent" not in root:
        for f in files:
            with open(os.path.join(root,f),'r',encoding='utf-8',errors='ignore') as e:
                data=e.read()
                email=Parser().parsestr(data)
                try: 
                    Rec = email["to"].split(',')
                    for r in Rec:
                        #r=r.replace(" ", "")
                        r=r.strip()
                        if r not in reciever:
                            reciever[r]=1
                        else:
                            reciever[r]+=1
                except (AttributeError):
                    pass
                    
```

```python
x1=[]
y1=[]
for x in sorted(sender, key=sender.get, reverse=True)[:25]:
    x1.append(x)
    y1.append(sender[x])
    print(x,sender[x])
```

```python
x2=[]
y2=[]
for x in sorted(reciever, key=reciever.get, reverse=True)[:25]:
    x2.append(x)
    y2.append(reciever[x])
    print(x,reciever[x])
```

```python
mp.style.use('ggplot') 
mp.style.use('seaborn-poster')
mp.subplot(111)
mp.bar(range(len(x1)),y1,align='center',color='plum')

mp.xticks(range(len(x1)),x1,rotation=90)
mp.ylabel('Emails Sent Count')
mp.xlabel('Employee Name')

mp.show()
```

```python
mp.style.use('ggplot') 
mp.style.use('seaborn-poster')
mp.bar(range(len(x2)),y2,align='center',color='gold')

mp.xticks(range(len(x2)),x2,rotation=90)
mp.ylabel('Emails Recieved Count')
mp.xlabel('Employee Name')

mp.show()
```
