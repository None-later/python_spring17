```python
from email.parser import Parser
import os,csv
import matplotlib.pyplot as mp
from datetime import datetime
import string, re
dir_path='/Users/nithinkartha/Desktop/Masters/study materials/DAUPy/HW/python_spring17/midterm/data/enron/maildir'
```

```python
mailTime=dict.fromkeys(range(0,24),0)
for root, dirs, files in os.walk(dir_path):
        if root.endswith("/sent"):
            for f in files:
                with open(os.path.join(root,f),'r',encoding='utf-8',errors='ignore') as e:
                    data=e.read()
                    email=Parser().parsestr(data)
                    
                    time=re.sub(':',' ',email["Date"][17:22])
                    h = int(time[0:2]) 
                    m = int(time[2:5])
                    if h in mailTime:
                        mailTime[h]+=1
                                  
```

```python
mp.style.use('ggplot')
mp.style.use('seaborn-dark-palette')
x=mailTime.keys()
y=mailTime.values()
mp.bar(range(len(mailTime)),y,align='center')
mp.xticks(range(len(mailTime)),x)

mp.show()
```

```python
MonthDay=dict.fromkeys(range(1,32),0)
for root, dirs, files in os.walk(dir_path):
        if root.endswith("/sent"):
            for f in files:
                with open(os.path.join(root,f),'r',encoding='utf-8',errors='ignore') as e:
                    data=e.read()
                    email=Parser().parsestr(data)
                    day=email["Date"][4:7]
                    day=int(day)
                    if day in MonthDay:
                        MonthDay[day]+=1
```

```python
x1=[]
y1=[]
for m in MonthDay:
    x1.append(m)
    y1.append(MonthDay[m])
```

```python
mp.style.use('ggplot') 
mp.style.use('seaborn-poster')
mp.bar(range(len(MonthDay)),y1,align='center')
mp.xticks(range(len(MonthDay)),x1,rotation='vertical')
mp.subplots_adjust(bottom=0.01)
mp.ylabel('Mail Count')
mp.xlabel('Day of the Month')

mp.show()
```
